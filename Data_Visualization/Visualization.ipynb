```python
import seaborn as sns
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = [15,5] #setting the size of the plot
```

```python
df = pd.read_csv('Dataset/stockprice_15_04_2019-25_10_2017.csv')
```

```python
df.head()
```

```python
df.count()
```

```python
df.dropna(axis=0,inplace=True)
```

```python
df.groupby(by='traded_companies').count().head()
```

```python
df['traded_companies'].value_counts().head()
```

**Visualizing data in terms of year**

```python
year = list()
for y in df['date']:
    year.append(y.split('/')[2])
```

```python
df['Year']=year
```

```python
df.head()
```

```python
x=df.groupby(by='Year').count()
```

```python
x
```

```python
plt.plot(x.index.values,x['no_of_transaction'])
plt.xlabel('Year')
plt.ylabel('Number of Transactions')
```

```python
y=df.groupby('traded_companies').count()
y.head()
```

```python
z=y.sort_values(by='no_of_transaction',ascending=False).head()
z
```

```python
names=list(z.index.values)
pos = np.arange(len(names))
trans=list(z['no_of_transaction'])
plt.bar(pos,trans)
plt.xticks(pos,names)
plt.xlabel('Name of Bank',fontsize=16)
plt.ylabel('Number of transactions',fontsize=16)
plt.ylim((1660,1725))
```
